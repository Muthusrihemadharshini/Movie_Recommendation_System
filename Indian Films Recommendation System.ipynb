```python
# Create 'tags' column in df2
df2['tags'] = df2['Year'] + df2['Genre'] + df2['Language'] + df2['Votes'] + df2['Rating(10)']

# Create new_df and copy the 'tags' column from df2
new_df = df2[['Movie Name', 'ID', 'tags']].copy() # Include 'tags' in the new DataFrame
# Apply the stems function to the 'tags' column
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
import nltk
from nltk.stem import PorterStemmer
from sklearn.feature_extraction.text import CountVectorizer

nltk.download('punkt') # Download punkt tokenizer if not already downloaded

ps = PorterStemmer()

def stems(text):
    T = []
    for i in text.split():
        T.append(ps.stem(i))
    return " ".join(T)

# Create 'tags' column by joining relevant text columns with spaces
df2['tags'] = df2['Genre'].astype(str) + ' ' + \
              df2['Language'].astype(str) + ' ' + \
              df2['Votes'].astype(str)

# Apply stemming to the 'tags' column
df2['tags'] = df2['tags'].apply(stems)

# Create new_df and copy the 'tags' column from df2
new_df = df2[['Movie Name', 'ID', 'tags']].copy()

# Initialize CountVectorizer with adjusted parameters
cv = CountVectorizer(max_features=5000, stop_words='english', token_pattern=r"(?u)\b\w+\b") # Adjust token_pattern

# Fit and transform the 'tags' data
vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
def recommend(movie):
  index = new_df[new_df['Movie Name'] == movie].index[0] # Changed 'title' to 'Movie Name'
  distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
  for i in distances[1:6]:
    print(new_df.iloc[i[0]]['Movie Name']) # Changed .Movie_Name to ['Movie Name'] for consistency
```
